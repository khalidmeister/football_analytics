```python
with open('open-data/data/events/2302764.json') as f:
    data = json.load(f)
    df = json_normalize(data, sep="_")

df.head()
```

```python
df[df.type_name == 'Half Start']
```

```python
first_half = df.loc[:1808]
second_half = df.loc[1809:3551]
second_half.tail()
```

# First Half

```python
block = first_half[first_half.type_name == 'Block']
block = block[['team_name', 'player_name', 'location']]

clearance = first_half[first_half.type_name == 'Clearance']
clearance = clearance[['team_name', 'player_name', 'location']]

intercept = first_half[first_half.type_name == 'Interception']
intercept = intercept[intercept.interception_outcome_name == 'Won']
intercept = intercept[['team_name', 'player_name', 'location']]

defense = pd.concat([block, clearance, intercept])

defense['x'] = defense.location.apply(lambda x: x[0])
defense['y'] = defense.location.apply(lambda x: x[1])
defense = defense.drop('location', axis=1)

defense.head()
```

```python
liverpool = defense[defense.team_name == 'Liverpool']

names = liverpool.player_name.unique()
colors = ['red', 'yellow', 'green', 'blue', 'purple', 'white', 'cyan', 'pink']

pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(8, 6))

for i, name in enumerate(names):
    data = liverpool[liverpool.player_name == name]
    if data.shape[0] > 2:
        hull = pitch.convexhull(data.x, data.y)
        poly = pitch.polygon(hull, ax=ax, edgecolor=colors[i], facecolor=colors[i], alpha=0.3)
    scatter = pitch.scatter(data.x, data.y, ax=ax, edgecolor='black', facecolor=colors[i])
    cen_x = np.sum(data.x) / data.shape[0]
    cen_y = np.sum(data.y) / data.shape[0]        
    pitch.annotate(name, xy=(cen_x, cen_y), c=colors[i], va='center', ha='center', size=12, ax=ax)

plt.show()
```
