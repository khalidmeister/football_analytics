```python
with open('open-data/data/events/2302764.json') as f:
    data = json.load(f)
    df = pd.json_normalize(data, sep="_")
    
df.head()
```

```python
df = df.loc[:4627]
shots = df[df.type_name == 'Shot']
shots = shots[['team_name', 'player_name', 'minute', 'second', 'location', 'shot_statsbomb_xg', 'shot_outcome_name']]
shots.head()
```

```python
h_xg = []
a_xg = []
h_min = []
a_min = []

for i, data in shots.iterrows():
    if data.team_name == 'AC Milan':
        h_xg.append(data.shot_statsbomb_xg)
        h_min.append(data.minute)
    else:
        a_xg.append(data.shot_statsbomb_xg)
        a_min.append(data.minute)
        

h_xg = np.array(h_xg)
a_xg = np.array(a_xg)
        
print(h_xg)
print(h_min)
print()
print(a_xg)
print(a_min)
```

```python
cum_h_xg = np.round(np.cumsum(h_xg), 2)
cum_a_xg = np.round(np.cumsum(a_xg), 2)

print(cum_h_xg)
print(cum_a_xg)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_alpha(0)
ax.spines['right'].set_alpha(0)

ax.tick_params(colors='white')

ax.step(x=h_min, y=cum_h_xg, label='AC Milan')
ax.step(x=a_min, y=cum_a_xg, label='Liverpool')

ax.set_xlabel('xG')
ax.set_ylabel('Minute')

ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

ax.set_title('The xG Progress Chart between Liverpool and AC Milan')
ax.title.set_color('white')

plt.legend()
plt.show()
```
